## Example 1: Running Sum of 1d Array

**Problem:**Given an array. Running sum of all preceding elements in that array along with that element.
**[LeetCode #1480](https://leetcode.com/problems/running-sum-of-1d-array/)**

```
python

````python
def running_sum_single_pass_traversal(nums):
    """
    Finding running sum of an array.

    Input: [2, 3, 6, 7, 9]
    Output: [2, 5, 11, 18, 27]
    Explanation: [2, 2+3, 2+3+6, 2+3+6+7, 2+3+6+7+9]

    Approach: Add all preceding elements upto that element in entire element
    """

    # if array is empty, no running sum for array
    if not nums:
        return None
 
    # one loop through array
    for i in range(1,len(nums)):
        # adding total sum upto last element + current element
        nums[i]=nums[i-1]+nums[i]
        

    return nums # list with running sum
````

````python
# Test
nums=[2, 3, 6, 7, 9]
print(running_sum_single_pass_traversal(nums))
````

## Example 2:

**Problem:** Number of 1 Bits.            ***[Leetcode #191](https://leetcode.com/problems/number-of-1-bits/)***

```
python

````python
def no_of_1bits_single_pass_traversal(n):
    """
    Finding the number of 1 in it's Binary Form

    Input: 19
    Output: 3
    Explination : binary(19) -> 10011 [3 1-Bits]

    Approach: Convert the interger into binary, store the bits in array , loop through the array once by maintaining the count of 1Bits
    """

    # Variable to count number of 1 Bits
    bits_count=0
    # List to story binary of numbers
    binary=[]
    # Loop to find and add binary numbers into the list
    while n>0:
        # adding the every binary number to list
        binary.append(n%2)
        n//=2


    # Traversing begins
    # Looping through the binary array once
    for bit in binary:
        # if bit=1 we'll Increase the count of bits
        if bit:
            bits_count+=1
    # Result of number of bits        
    return bits_count        


"""
### Time Complexity:

- While loop: The loop runs until n becomes 0. Since in each iteration, n is halved, this loop runs in O(log n) time, where n is the number itself (since dividing by 2 reduces the size of n).

For loop: The second loop goes through each bit in the binary list. This loop also runs for O(log n) times because there are at most log n bits in the binary representation.

Overall Time Complexity: O(log n) due to both loops.

### Space Complexity:

- The space complexity is O(log n) because you are storing the binary representation in the binary list, and the number of bits is proportional to the logarithm of the number.
"""
````

````python
# Test
number=19
print(no_of_1bits_single_pass_traversal(number))
# 3
````
