## Example 1: Find the Middle index of the Array 

### Problem: find the index we the sum of elements before that index is equal to sum of elements after that index      excluding that index [LeetCode 1991](https://leetcode.com/problems/find-the-middle-index-in-array)

python


```python
def find_middle_index(nums):
    """
    Finding the middle index

    Input: [2, 3, -1, 8, 4]
    Output: 3
    Explaination: 2+3+(-1) = 4 

    Approach: For every element check left sum and right sum
    """

    # if array is empty, no middle index
    if not nums:
        return None

    # Taking total sum to calculate Right sum at every element
    total_sum = sum(nums)
    # Left sum , starting with zero because not left elements for 1st element
    left_sum = 0

    # one loop through the array
    for i in range(len(nums)):
        # calculation of right_sum for that index element
        right_sum = total_sum-left_sum-nums[i] # that element will not include

        # condition for middle index
        if right_sum == left_sum:
            return i
        # incrementing left_sum after right sum calculation
        left_sum += nums[i]

    # returns -1 , if there is no middle index
    return -1       
```

```python
# Test
find_middle_index([2, 3, -1, 8, 4])
```

* Time Complexity: O(n)
* Space Complexity: O(1)
