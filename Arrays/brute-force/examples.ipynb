## Example 1: Two Sum (Classic)

**Problem:** Given array and target, find two numbers that add up to target.

```
python

````python
def two_sum_brute_force(nums,target):
    """
    Find two numbers that sum to target.

    Input: nums=[2,7,11,15], target = 9
    Output: [0,1]
    Explanation: nums[0]+nums[1] = 2 + 7 = 9

    Approach: Try every pair
    """

    n=len(nums)

    #Try first number
    for i in range(n):
        #Try second number
        for j in range(i+1,n):
            #Check if they sum to target
            if nums[i]+nums[j]==target:
                return [i,j]

    return None # if not solution found           
````

````python
# Test
nums,target=[2,7,11,15],9
print(two_sum_brute_force(nums,target))
# [0, 1]
````

---
**Dry Run:**

nums=[2, 7, 11, 15], target=9

i=0 , nums[0] = 2:
  j=1, nums[1] = 7: 2+7= 9 Found! Return [0,1]

Result: [0, 1]
```

**Time:**O(n²) - nested loop
**Space:**O(1) - only variables

def find_max_brute_force(arr):
    """
    Find the maximum element.

    Input: arr=[3, 1, 4, 1, 5, 9, 2]
    Output: 9

    Approach: Compare every element with every other!
    (Not optimal, but demonstrates brute force)
    """
    if not arr: # if no element in array, return's None
        return None

    n=len(arr)
    max_value=arr[0]

    #check each element
    for i in range(n):
        #Against all elements
        is_max=True
        for j in range(n):
            if arr[j]>arr[i]:
                is_max=False
                break
        if is_max:
            max_val=arr[i]

    return max_val

````python
# Test
arr = [3, 1, 4, 1, 5, 9, 2]
print(find_max_brute_force(arr))
# 9
````

**Note** This is unnecessarily complex!

**Better approach:** Single loop O(n):
```
python

````python
def find_max_optimized(arr):
    max_val=arr[0]
    for num in arr:
        max_val=max(max_val,num)
    return max_val
````
